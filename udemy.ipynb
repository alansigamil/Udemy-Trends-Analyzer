```python
df = pd.read_csv('udemy_courses.csv')  

def func_draw(x):
    # Bar plot
    plt.figure(figsize=(20, 10))
    sns.barplot(x='instructional_level', y=x, data=df, hue='category')
    plt.title('Plot of {} per Instructional Level per Category'.format(x), fontsize=16)
    plt.xlabel('Instructional Level', fontsize=14)
    plt.ylabel(x, fontsize=14)
    plt.legend(title='Category', fontsize=12)
    plt.show()

    # Pie chart
    plt.figure(figsize=(10, 10))
    pie_data = df.groupby('instructional_level')[x].sum()
    pie_data.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='Set3')
    plt.title('Plot of {} per Instructional Level'.format(x), fontsize=16)
    plt.ylabel('')  
    plt.show()

func_draw('num_reviews')
```
